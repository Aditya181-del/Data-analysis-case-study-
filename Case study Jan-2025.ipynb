```python
#What is the overall price trend for each commodity ?
# Line chart to check the trends over time. 
fig = plt.subplots(2,2,figsize = (20,10))
plt.subplot(2,2,1)
plt.plot(df['coffee_arabica'],label = 'coffee robustas price',color = 'Blue')
plt.title("Coffee Arabica trend")
plt.ylabel("Price")
plt.xlabel("Time")
plt.legend()


plt.subplot(2,2,2)
plt.plot(df['coffee_robustas'],label = 'coffee dubai price',color = 'Red')
plt.title("Coffee Robustas trend")
plt.ylabel("Price")
plt.xlabel("Time")
plt.legend()


plt.subplot(2,2,3)
plt.plot(df['oil_brent'],label = 'Oil brent price',color = 'Orange')
plt.title("Oil brent trend")
plt.ylabel("Price")
plt.xlabel("Time")
plt.legend()

plt.subplot(2,2,4)
plt.plot(df['oil_dubai'],label = 'Oil dubai price',color = 'Green')
plt.title("Oil dubai trend")
plt.ylabel("Price")
plt.xlabel("Time")
plt.legend()
plt.show()
```

```python
#Hist plots for checking price distribution.
fig = plt.subplots(2,2,figsize = (20,10))
plt.subplot(2,2,1)
plt.hist(df['tea_columbo'],color = 'goldenrod',bins=30,edgecolor = 'black')
plt.title("Tea Columbo price distribution")
plt.ylabel("Price")
plt.xlabel("Time")

plt.subplot(2,2,2)
plt.hist(df['tea_kolkata'],color = 'Grey',bins=30,edgecolor = 'black')
plt.title("Tea Kolkata price distribution")
plt.ylabel("Price")
plt.xlabel("Time")

plt.subplot(2,2,3)
plt.hist(df['tea_mombasa'],color = '#3498DB',bins=30,edgecolor = 'black')
plt.title("Tea Mombasa price distribution")
plt.ylabel("Price")
plt.xlabel("Time")

plt.subplot(2,2,4)
plt.hist(df['sugar_eu'],color = 'maroon',bins=30,edgecolor = 'black')
plt.title("Sugar Europe price distribution")
plt.ylabel("Price")
plt.xlabel("Time")
plt.show()
```

```python
#Line plots to check the time series of USA prices
fig = plt.subplots(2,2,figsize = (20,10))
plt.subplot(2,2,1)
sns.lineplot(data=df,x='date', y='sugar_us', color='#a52a2a')
plt.title('Time Series of Sugar USA Prices')

plt.subplot(2,2,2)
sns.lineplot(data=df,x='date', y='sugar_world', color='#2e8b57')
plt.title('Time Series of Sugar World Prices')

plt.subplot(2,2,3)
sns.lineplot(data=df,x='date', y='coffee_arabica', color='#ffa500')
plt.title('Time Series of Coffee Arabica Prices')

plt.subplot(2,2,4)
sns.lineplot(data=df,x='date', y='oil_brent', color='#00008b')
plt.title('Time Series of Oil brent Prices')
plt.show()
```
